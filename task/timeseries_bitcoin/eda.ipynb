```python
import kagglehub
# Download latest version
path = kagglehub.dataset_download("mczielinski/bitcoin-historical-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

files = os.listdir(path)
data = pd.read_csv(f"{path}/{files[0]}")
data.head()
```

# 칼럼명 변경

```python
data.columns = ['timestamp', 'open', 'high', 'low', 'close', 'volume']
data.columns
```

# timestamp로 변환

```python
data['timestamp'].iloc[0].astype(str)
#10자리 유닉스 타임스탬프(초 단위, seconds)
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], unit='s')
```

```python
#인덱스 처리
data = data.set_index('timestamp', drop = False)
data.index
```

# EDA

### Dataset Column Description

- **Timestamp**  
  Start time of the time window (60s window), in Unix time

- **Open**  
  Open price at the start of the time window

- **High**  
  High price at the start of the time window

- **Low**  
  Low price at the start of the time window

- **Close**  
  Close price at the start of the time window

- **Volume**  
  Volume of BTC transacted within this time window
