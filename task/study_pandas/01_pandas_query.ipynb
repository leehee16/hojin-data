```python
import pandas as pd
```

```python
# 데이터 생성
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 28, 32],
    'city': ['Seoul', 'Busan', 'Seoul', 'Incheon', 'Busan'],
    'salary': [50000, 60000, 75000, 55000, 70000]
}
df = pd.DataFrame(data)
```

```python
# 평균나이 구하기
df[df['city']=='Seoul'][['age']].agg('mean')
```

```python
# 나이가 30세 이상이고, 연봉이 60000 이상인 사람의 이름을 출력
df[(df['age']>=30) & (df['salary']>=60000)]['name']
```

```python
df.groupby('city')['salary'].agg('mean')
```
